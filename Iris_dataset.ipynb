```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df.head(5)
```

```python
# Plot distribution of sepal length
plt.figure(figsize=(10, 6))
sns.histplot(df['sepal_length'], kde=True, bins=20)
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot distribution of sepal length
plt.figure(figsize=(10, 6))
sns.histplot(df['sepal_width'], kde=True, bins=20)
plt.title('Distribution of Sepal_lwidth')
plt.xlabel('Sepal_width (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot distribution of sepal length
plt.figure(figsize=(10, 6))
sns.histplot(df['petal_length'], kde=True, bins=20)
plt.title('Distribution of Petal length')
plt.xlabel('Petal length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.pairplot(df, hue='species')
plt.title('Pair Plot of Iris Dataset')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='sepal_length', data=df)
plt.title('Box Plot of Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='petal_length', data=df)
plt.title('Box Plot of Petal length by Species')
plt.xlabel('Species')
plt.ylabel('Petal length (cm)')
plt.show()
```

```python
# Violin plot of sepal width by species
plt.figure(figsize=(10, 6))
sns.violinplot(x='species', y='sepal_width', data=df)
plt.title('Violin Plot of Sepal Width by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
# Violin plot of sepal width by species
plt.figure(figsize=(10, 6))
sns.violinplot(x='species', y='petal_width', data=df)
plt.title('Violin Plot of Petal Width by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
# Compute the correlation matrix
correlation_matrix = df.drop('species', axis=1).corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
